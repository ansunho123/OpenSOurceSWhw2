```python
import pandas as pd

file_path = '2019_kbo_for_kaggle_v2.csv'


data = pd.read_csv(file_path)

data2 = data[['batter_name','H','avg','HR','OBP','year']]
data2 = data2[(data2['year'] >= 2015) & (data2['year'] <= 2018)]
for year in range(2015,2019):
    ydata = data2[data2['year'] == year]
    print(year)
    print('\n')
    
    for standard in ['H','avg','HR','OBP']:
        temp = ydata[['batter_name',standard]]
        print(temp.sort_values(by=standard, ascending=False)[:10])
        
        
    print('\n')
    
data2_2 = data[['batter_name','war','cp','year']]
data2_2 = data2_2[data2_2['year'] == 2018]
mask = data2_2.groupby('cp')['war'].idxmax()
data2_2 = data2_2.loc[mask]
print(2018)
print(data2_2[['cp', 'batter_name']])




data2_3 = data[['R','H','HR','RBI','SB','war','avg','OBP','SLG']]
corrdata = data2_3.corrwith(data.salary)
print(corrdata)
print('\n')
print("highest correlation is : ", corrdata.idxmax())
```
